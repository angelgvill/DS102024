### Paso 4. ¿Qué continente bebe más cerveza por término medio?

```python
bebidas['continent'].unique()
```

```python
bebidas[bebidas['continent'].isnull()
```

```python
bebidas[bebidas['country']== 'Mexico']  #veo aquí que paises como Mexico no están asignados a continente
```

```python
bebidas['continent'].fillna('NA', inplace=True)
```

```python
bebidas.groupby('continent')[['beer_servings']].mean().sort_values('beer_servings', ascending=False)
```

### Paso 5. Para cada continente, imprima las estadísticas de consumo de vino.

```python
bebidas.groupby('continent')[['wine_servings']].describe()
```

### Paso 6. Imprime la media de consumo de alcohol por continente para cada columna

```python
bebidas.groupby('continent')[['total_litres_of_pure_alcohol']].mean() #y añado cada columna con comas
```

```python
bebidas.groupby('continent').mean(numeric_only=True)
```

### Paso 7. Imprime la mediana de consumo de alcohol por continente para cada columna

```python
bebidas.groupby('continent').median(numeric_only=True)
```

### Paso 8. Imprima los valores medio, mínimo y máximo de consumo de alcohol por continente.

```python
bebidas.groupby('continent')[['total_litres_of_pure_alcohol']].agg(['min', 'max', 'median'])
```
