# Nombre del producto de valor máximo

```python
chipo.info() #vemos que item price es un objeto, no un float
```

```python
chipo.describe()
```

```python
#por el profe
chipo['item_price'] = chipo['item_price'].str.replace('$',' ')
```

```python
#opción por profe
chipo['item_price'] = chipo['item_price'].str.strip().astype('float')
chipo
```

```python
chipo['item_price_float'] = chipo['item_price']     #replico la columna para transformarla
chipo['item_price_float'] = chipo['item_price_float'].str.strip().str.slice(start=1)

chipo['item_price_float'] = chipo['item_price_float'].astype('float32') #cambio el string por un flotante
chipo.head()
```

```python
chipo['item_price_float'].info()
```

```python
max_price_item = chipo.loc[chipo['item_price_float'].idxmax(), 'item_name']

chipo['item_price_float'].max()
chipo['item_price_float'].idxmax()
print (f'el item con el precio más alto es {max_price_item}, con un precio de {chipo['item_price_float'].max()}, en la posición {chipo['item_price_float'].idxmax()}')
```

```python
#si hacemos por el precio del item individual
chipo['unit_price'] = chipo['item_price_float'] / chipo['quantity']

# Localizado el índice del valor unitario más alto
max_unit_price_idx = chipo['unit_price'].idxmax()

chipo.sort_values('unit_price', ascending=False)
```

```python
#aquí mostramos el item con el precio unitario más alto
most_expensive_item = chipo.loc[max_unit_price_idx, 'item_name']
most_expensive_item  #hay muchos, entiendo toma la primera aparición o así.
```

```python
#con Grouby, agrupaciones
chipo.groupby('item_name')['item_price_float'].max()
```

```python
#aún no estaría la opción del profe, continuamos
cual = chipo.groupby('item_name')['item_price'].max()
cual[cual['quantity'] == 1].sort_values('item_price',ascending = False).head(1)
```
