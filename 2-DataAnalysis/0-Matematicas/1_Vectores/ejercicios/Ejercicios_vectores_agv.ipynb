## Ejercicio 1: Ortogonalidad

Definir una función en python que detecte si dos vectores son ortogonales

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def vector_norm(vector):
    squares = [element**2 for element in vector]
    return sum(squares)**0.5

# print("||", u, "|| =")
# vector_norm(u)
```

```python
import numpy as np
def son_ortogonales(u,v):
  #completa con tu código aquí
  if vector_norm(u) == 0:
    return (f'vector u norma cero')
  if vector_norm(v) == 0:
    return (f'vector v norma cero')
  if np.dot(u,v) == 0:
    return (f'Los vectores son ortogonales (producto escalar cero)')
  else:
    return (f'Los vectores no son orgononales (producto escalar no nulo), {np.dot(u,v)}')

u = np.array([0, 9])
v = np.array([-1, 1])
print("||", u, "|| =",vector_norm(u) )


print("||", v, "|| =", vector_norm(v))


son_ortogonales (u,v)
```

## Ejercicio 2: Desplazamiento

Definir una función en python que reciba como parámetro las cuatro esquinas de un cuadrilátero y las desplace 5 unidades hacia arriba y 7 unidades a la izquierda

```python
def desplazar_puntos(a,b,c,d):
  #completá con tu código aquí
  v = np.array([-7,5])
  a_ = a + v
  b_ = b + v
  c_ = c + v
  d_ = d + v
  print (f'tu lista de coordenadas de las esquinas es {a,b,c,d}')
  print (f'tus matrices de coordenadas desplazadas son {a_,b_,c_,d_}')
  print (f'tus matrices de coordenadas desplazadas son {a_},{b_},{c_},{d_}')
  return (a_,b_,c_,d_)
```

```python
desplazar_puntos ([2, 2],[6,7],[3,12],[7,12])
```

```python
def plot_vector2d(vector2d:np.array, origin:list=[0, 0], **options) -> None:
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1],
              head_width=.5, head_length=.5, length_includes_head=True,
              **options)
```

```python
t1 = np.array([2, 2])
t2 = np.array([6, 7])
t3 = np.array([7, 12])
t4 = np.array([3, 12])

v = np.array([-7,5])

print(f'voy a dibujarte el movimiento en 2D para que quede más claro:')

x_coords, y_coords = zip(t1, t2, t3, t4, t1)
plt.plot(x_coords, y_coords, "c--", x_coords, y_coords, "co")

plot_vector2d(v, t1, color="r", linestyle=":")
plot_vector2d(v, t2, color="r", linestyle=":")
plot_vector2d(v, t3, color="r", linestyle=":")
plot_vector2d(v, t4, color="r", linestyle=":")

t1b = t1 + v
t2b = t2 + v
t3b = t3 + v
t4b = t4 + v
(t1b, t2b, t3b, t4b) =desplazar_puntos(t1,t2, t3,t4)

x_coords_b, y_coords_b = zip(t1b, t2b, t3b, t4b, t1b)
plt.plot(x_coords_b, y_coords_b, "b-", x_coords_b, y_coords_b, "bo")

plt.text(-1, 2.8, "v", color="r", fontsize=18)
plt.text(1, 9, "v", color="r", fontsize=18)
plt.text(.5, 12.5, "v", color="r", fontsize=18)
plt.text(4, 14.8, "v", color="r", fontsize=18)

plt.axis([-9, 9, 0, 18])
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
a = np.array([1,1])
b = np.array([1,2])
c = np.array([2,3])
d = np.array([8,4])

x_coords, y_coords = zip(a,b,c,d,a)
plt.plot(x_coords, y_coords, "c--", x_coords, y_coords, "co")

a_,b_,c_,d_ = desplazar_puntos(a,b,c,d)
x_coords_b, y_coords_b = zip(a_,b_,c_,d_,a_)
plt.plot(x_coords_b, y_coords_b, "b-", x_coords_b, y_coords_b, "bo")

plt.show()
```
